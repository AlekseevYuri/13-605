С выключенными двигателями и без сопротивления воздуха ракета движется как обычное тело - по параболе.

```python
import math

MODEL_G = 9.81
MODEL_DT = 0.001
U = 200
mu = 0
gamma = 0
k = 0
class Body:
    def __init__(self, x, y, vx, vy, m):
        """
        Создать тело.
        
        Пареметры:
        ----------
        x: float
            горизонтальная координата
        y: float
            вертикальная координата
        vx: float
            горизонтальная скорость
        vy: float
            вертикальная скорость
        """

        self.x = x
        self.y = y
        self.vx = vx
        self.vy = vy
        self.m  = m
        
        self.trajectory_x = []
        self.trajectory_y = []
        

    def advance(self):
        """
        Выполнить шаг мат. модели применительно к телу, предварительно записав его координаты
        """
        self.trajectory_x.append(self.x)
        self.trajectory_y.append(self.y)
        
        self.x += self.vx * MODEL_DT
        self.y += self.vy * MODEL_DT
        self.vy -= MODEL_G * MODEL_DT

 
        
        
class Rocket2(Body):
    def __init__(self, x, y):
        super().__init__(x, y, 9, 9,1000) # Вызовем конструктор предка — тела, т.к. он для ракеты актуален
        pass # Дописать
    def advance(self):
        super().advance() # вызовем метод предка — тела, т.к. и он для ракеты актуален.
        pass # Дописать
        if self.m > 200:           
            self.vx +=  self.vx/(math.sqrt(self.vy**2+self.vx**2)) * U * mu / self.m-gamma * self.vx
            self.vy +=  self.vy/(math.sqrt(self.vy**2+self.vx**2)) * U * mu / self.m-gamma * self.vy
            self.m -= mu
            if self.m < 950:
                if k == 1:
                    self.m-=500
import numpy as np

np.sin

b = Body(0, 0, 9, 9,1000)
r = Rocket2(0, 0)

bodies = [b, r]
for t in np.r_[0:2:MODEL_DT]: # для всех временных отрезков
    for b in bodies: # для всех тел
        b.advance() # выполним шаг
%matplotlib inline
from matplotlib import pyplot as pp

for b in bodies: # для всех тел
    pp.plot(b.trajectory_x, b.trajectory_y) # нарисуем их траектории
```
